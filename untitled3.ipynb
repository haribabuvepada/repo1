```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (assuming the CSV file is named 'water_quality.csv')
# If the file name is different or if it's not in the current directory,
# please update the path accordingly.
try:
    df = pd.read_csv('water_quality.csv')
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("Error: 'water_quality.csv' not found. Please upload the file or provide the correct path.")
    # Create a dummy DataFrame for demonstration if the file is not found
    data = {
        'ph': np.random.uniform(6.0, 9.0, 1000),
        'Hardness': np.random.uniform(100, 300, 1000),
        'Solids': np.random.uniform(200, 50000, 1000),
        'Chloramines': np.random.uniform(0.5, 12, 1000),
        'Sulfate': np.random.uniform(200, 400, 1000),
        'Conductivity': np.random.uniform(200, 700, 1000),
        'Organic_carbon': np.random.uniform(2, 25, 1000),
        'Trihalomethanes': np.random.uniform(50, 120, 1000),
        'Turbidity': np.random.uniform(1, 6, 1000),
        'Potability': np.random.randint(0, 2, 1000)
    }
    df = pd.DataFrame(data)
    # Introduce some missing values for demonstration
    for col in ['ph', 'Sulfate', 'Trihalomethanes']:
        df.loc[df.sample(frac=0.1).index, col] = np.nan
    print("Generated a dummy DataFrame with simulated missing values for demonstration.")

# Display the first few rows of the dataset
display(df.head())

# Get a summary of the dataset
display(df.info())
```

### EDA: Distributions of pH, Hardness, and Sulfate

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['ph'].dropna(), kde=True)
plt.title('Distribution of pH')

plt.subplot(1, 3, 2)
sns.histplot(df['Hardness'], kde=True)
plt.title('Distribution of Hardness')

plt.subplot(1, 3, 3)
sns.histplot(df['Sulfate'].dropna(), kde=True)
plt.title('Distribution of Sulfate')

plt.tight_layout()
plt.show()
```

### EDA: Missing Value Pattern

```python
print("Missing values per column:")
display(df.isnull().sum())

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Value Pattern')
plt.show()
```
